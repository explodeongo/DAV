```python
#g

# Function to remove outliers based on IQR
def remove_outliers(df, column, threshold=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    
    # Identify outliers
    outliers = (df[column] < lower_bound) | (df[column] > upper_bound)
    
    # Remove outliers
    df_no_outliers = df[~outliers]
    
    return df_no_outliers

# Remove outliers from each column
for column in df.columns:
    df = remove_outliers(df, column)

# Display the result
print("Original DataFrame:")
print(df)
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Create sample data for day 1
data_day1 = {
    'Name': ['John', 'Alice', 'Bob', 'Charlie'],
    'Time of joining': [datetime(2023, 1, 1, 9, 0), datetime(2023, 1, 1, 9, 15), datetime(2023, 1, 1, 9, 30), datetime(2023, 1, 1, 10, 0)],
    'Duration': [30, 40, 50, 30]
}

df_day1 = pd.DataFrame(data_day1)

# Create sample data for day 2
data_day2 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Time of joining': [datetime(2023, 1, 2, 9, 10), datetime(2023, 1, 2, 9, 30), datetime(2023, 1, 2, 10, 0), datetime(2023, 1, 2, 9, 15)],
    'Duration': [40, 50, 30, 40]
}

df_day2 = pd.DataFrame(data_day2)

# Save data to CSV files
df_day1.to_csv('day1.csv', index=False)
df_day2.to_csv('day2.csv', index=False)
```

```python
df_day1 = pd.read_csv('day1.csv')
df_day2 = pd.read_csv('day2.csv')
```

```python
# a. Perform merging of the two dataframes to find the names of students who had attended the workshop on both days.
common_names = pd.merge(df_day1, df_day2, on='Name', how='inner')['Name']
print("Names of students who attended the workshop on both days:")
print(common_names)

# b. Find names of all students who have attended workshop on either of the days.
all_names = pd.merge(df_day1, df_day2, on='Name', how='outer')['Name']
print("\nNames of all students who attended the workshop on either of the days:")
print(all_names)

# c. Merge two data frames row-wise and find the total number of records in the data frame.
merged_df = pd.concat([df_day1, df_day2], ignore_index=True)
total_records = len(merged_df)
print(f"\nTotal number of records in the merged data frame: {total_records}")

# d. Merge two data frames and use two columns names and duration as multi-row indexes. Generate descriptive statistics for this multi-index.
merged_multiindex_df = pd.merge(df_day1, df_day2, on=['Name', 'Duration'], how='inner')
stats = merged_multiindex_df.groupby(['Name', 'Duration']).describe()
print("\nDescriptive statistics for the multi-index:")
print(stats)
```
